The delivery_time.csv dataset contains information about Delivery time and Sorting Time. We have to prepare linear regression and 
polynomial regression model to predict Sorting time on the basis of Delivery Time. Also evaluate the model by using evaluation metrics.

```python
import warnings 
warnings.filterwarnings(action="ignore")
```

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load dataset
df = pd.read_csv("dataset/delivery_time.csv")
```

```python
# Rename columns if necessary
df.columns = ['Delivery_Time', 'Sorting_Time']
```

```python
df.head()
```

<h1>Linear Regression</h1>


```python
# Features and target
X = df[['Delivery_Time']]
y = df['Sorting_Time']
```

```python
# Linear Regression
linear_modell  = LinearRegression()
linear_modell.fit(X, y)
y_pred_linear = linear_modell.predict(X)
```

```python
# Evaluation
print("\nLinear Regression Evaluation:")
print("Mean Squared Error:", mean_squared_error(y, y_pred_linear))
print("R² Score:", r2_score(y, y_pred_linear))
```

```python

# Plotting
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred_linear, color='green', label='Linear Prediction')
plt.xlabel('Delivery Time')
plt.ylabel('Sorting Time')
plt.legend()
plt.title('Linear Regression')
plt.show()
```

<h1>Polynomial Regression</h1>


```python

# Polynomial Regression (degree=2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

poly_model = LinearRegression()
poly_model.fit(X_poly, y)
y_pred_poly = poly_model.predict(X_poly)

# Evaluation


print("\nPolynomial Regression Evaluation:")
print("Mean Squared Error:", mean_squared_error(y, y_pred_poly))
print("R² Score:", r2_score(y, y_pred_poly))

```

```python
# Plotting
plt.scatter(X, y, color='blue', label='Actual')

# Sort X for smooth curve
sorted_zip = sorted(zip(X['Delivery_Time'], y_pred_poly))
X_sorted, y_poly_sorted = zip(*sorted_zip)

plt.plot(X_sorted, y_poly_sorted, color='red', label='Polynomial Prediction')
plt.xlabel('Delivery Time')
plt.ylabel('Sorting Time')
plt.legend()
plt.title('Polynomial Regression (Degree 2)')
plt.show()
```
